```python

records = load_data('./data/clean_data.jsonl')

fig = plt.figure(figsize=(18, 12))
gs = GridSpec(ncols=2, nrows=3, figure=fig, 
              height_ratios=[1, 1, 1],  
              width_ratios=[1, 2])  

# a)
ax0 = fig.add_subplot(gs[0,0])
plot_log_log_entities(records, ax=ax0)
ax0.text(-0.1, 1.05, 'a)', transform=ax0.transAxes, 
         size=18, va='top', ha='right')

# b)
ax1 = fig.add_subplot(gs[1,0])
plot_leader_counts_by_language(records, ax=ax1)
ax1.text(-0.1, 1.05, 'b)', transform=ax1.transAxes, 
         size=18, va='top', ha='right')

# c)
ax2 = fig.add_subplot(gs[2,0])
plot_word_count_distribution(records, ax=ax2)
ax2.text(-0.1, 1.05, 'b)', transform=ax2.transAxes, 
         size=18, va='top', ha='right')

# d)
ax_img = fig.add_subplot(gs[:,1])
img = Image.open('./data/person-person.png')
ax_img.imshow(img, aspect='auto')
ax_img.text(-0.01, 1.023, 'd)', transform=ax_img.transAxes, 
            size=18, va='top', ha='right')
ax_img.axis('off')

plt.tight_layout()
# plt.savefig("./some_path.pdf", 
#             dpi=300, 
#             bbox_inches='tight')
plt.show()
```
